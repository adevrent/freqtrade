```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import find_peaks
```

```python
x = np.linspace(0, 10, 100)
y = x * np.random.randn(100)**2
x
```

```python
y
```

```python
plt.plot(y)
```

```python
peaks_y = find_peaks(y, distance = 3)
peaks_y  # Whole output with dictionary being empty as second element.
```

```python
peaks_y[0]  # Peak indexes as the first value of the output.
```

```python
y[peaks_y[0]]  # Peak values.
```

```python
plt.plot(y)

plt.plot(y[peaks_y[0]])

# Should make their indexes equal to make a correct plot. To do.
```
